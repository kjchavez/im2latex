```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
from util.data_utils import *
import zmq
```

```python
class InferenceClient(object):
    def __init__(self, address):
        self.sock = zmq.Context.instance().socket(zmq.REQ)
        self.sock.connect(address)
        
    def predict(self, filename):
        self.sock.send(filename)
        return self.sock.recv()
    
client = InferenceClient('tcp://127.0.0.1:12345')
```

```python
image_filename = "data/cropped_images/227301e73d.png"
visualize_formula(image_filename)

# Make sure the inference server is running before making these calls.
formula = client.predict(image_filename)
print formula
print "Length:", len(formula)
```
